```python
#Código que gera base fictícia

import pandas as pd
import numpy as np
from random import choice, randint

# Leio os dados obtidos através de pesquisas das planilhas que criamos e as coloco em um dataframe
#df_carteira = pd.read_csv('/content/drive/MyDrive/Embraer/Carteira_de_pedidos.csv')
#df_avioes = pd.read_csv('/content/drive/MyDrive/Embraer/Avioes_entregues.csv')

#Lendo dados do github
df_carteira = pd.read_csv('https://github.com/NatashaB-randao/CustoDeTransporte-EmbraerSocialTechCareers/blob/main/Databases%20to%20complete%20missing%20values/Carteira_de_pedidos.csv')
df_avioes = pd.read_csv('https://github.com/NatashaB-randao/CustoDeTransporte-EmbraerSocialTechCareers/blob/main/Databases%20to%20complete%20missing%20values/Avioes_entregues.csv')
```

```python
# Declaro uma variável sgemtno com a coluna 'Aviação' da base aviões
segmento = df_avioes['Aviação'].to_list()
#segmento
```

```python
# Declaro uma variável com os modelo do avião
modeloAviao = df_avioes['Modelo do Avião'].to_list()
#modeloAviao
```

```python
# Declaro uma variável com os anos de entrega
ano_entrega = df_avioes['Ano de entrega'].to_list()
#ano_entrega
```

```python
#Declaro uma variável com a quantidade de aviões entregue
qtd_entregue = df_avioes['Quantidade entregue'].to_list()
#qtd_entregue
```

```python
''' Aqui gero aleatoriamente os paisese e cias aéreas para completar
nossa base de dados. Além disso, também repartimos as quantidades entregues
em diferentes países, não tinhamos os paises aos quais aeronaves foram
entregues, porém sabemos que elas cobrem o período de 2013 a 2023'''

lista_paises = []
lista_cia = []
lista_segmento = []
lista_modeloAviao = []
lista_qtdEntregue = []
lista_anoEntrega = []

for i in range(100):
  aviao = modeloAviao[i]
  seg = segmento[i]
  ano = ano_entrega[i]
  qtd = qtd_entregue[i]
  while qtd != 0:
    entrega = randint(1,qtd)
    qtd = qtd - entrega
    lista_qtdEntregue.append(entrega)
    #print(qtd)
    lista_modeloAviao.append(aviao)
    lista_anoEntrega.append(ano)
    lista_segmento.append(seg)
    #escolho aleatoriamente um pais da lista paises
    pais = choice(paises)
    #coloco os valores aleatorios em uma lista
    lista_paises.append(pais)
    #print(pais)
    #escolho aleatoriamente uma cia aerea que corresponde ao pais escolhido anteriormente
    cia_aerea = choice(df_carteira[df_carteira['País'] == pais]['Cia Aérea'].unique())
    #coloco os valores de cia aerea em uma lista
    #print(cia_aerea)
    lista_cia.append(cia_aerea)


print(np.shape(lista_paises))
print(np.shape(lista_cia))
print(np.shape(lista_anoEntrega))
print(np.shape(lista_modeloAviao))
print(np.shape(lista_segmento))
print(np.shape(lista_qtdEntregue))


#faço um dicionário com os valores gerados aleatoriamente
dados_random = {
    'Segmento': lista_segmento,
    'Modelo do Avião': lista_modeloAviao,
    'Quantidade Entregue': lista_qtdEntregue,
    'Ano de Entrega': lista_anoEntrega,
    'País': lista_paises,
    'Cia Aérea': lista_cia
}

#transformo o dicionário em um dataframe
df_random_paisECia = pd.DataFrame(dados_random)
```

```python
#para gerar uma lista precisamos inicializa com valor nulo
lista_modelo = []
lista_ano = []
lista_entrega = []
lista_Pais = []
lista_Cia_aerea = []
lista_pedidosFirmes = []

'''gero aleatoriamente os anos de acordo com o modelo do aviao (dentro do ano de fabricação de cada avião)
gero também aleatoriamente as quantidades entregues em diferentes anos '''
for i in range(0,129):
  modelo = modelo_aviao[i]
  pais_var = pais_augmentation[i]
  cia = cia_aerea_aug[i]
  pedidos = pedidos_carteira[i]
  cont = 0
  #lista_modelo.append(modelo)
  var = entregas[i]
  #cont = var
  if modelo == 'EMBRAER 170 (E170)':
    while var != 0:
      entrega = randint(1,var)
      var = var - entrega
      lista_entrega.append(entrega)
      #print(var)
      lista_modelo.append(modelo)
      # se o pais não consta como divulgado, gero aleatoriamente um pais em seu lugar
      if pais_var == 'Clientes não divulgados':
        excecao = choice(pais_augmentation)
        lista_Pais.append(excecao)
        lista_Cia_aerea.append(choice(df_carteira[df_carteira['País'] == excecao]['Cia Aérea'].unique()))
      else:
        lista_Pais.append(pais_var)
        lista_Cia_aerea.append(cia)
      if cont == 0:
        lista_pedidosFirmes.append(pedidos)
        cont = cont + 1
      else:
        lista_pedidosFirmes.append(0)
      ano = choice(e170)
      lista_ano.append(ano)

  elif modelo == 'EMBRAER 175 (E175)':
    while var != 0:
      entrega = randint(1,var)
      var = var - entrega
      lista_entrega.append(entrega)
      #print(var)
      lista_modelo.append(modelo)
      if pais_var == 'Clientes não divulgados':
        excecao = choice(pais_augmentation)
        lista_Pais.append(excecao)
        lista_Cia_aerea.append(choice(df_carteira[df_carteira['País'] == excecao]['Cia Aérea'].unique()))
      else:
        lista_Pais.append(pais_var)
        lista_Cia_aerea.append(cia)
      if cont == 0:
        lista_pedidosFirmes.append(pedidos)
        cont = cont + 1
      else:
        lista_pedidosFirmes.append(0)
      ano = choice(e175)
      lista_ano.append(ano)

  elif modelo == 'EMBRAER 190 (E190)':
    while var != 0:
      entrega = randint(1,var)
      var = var - entrega
      lista_entrega.append(entrega)
      #print(var)
      lista_modelo.append(modelo)
      if pais_var == 'Clientes não divulgados':
        excecao = choice(pais_augmentation)
        lista_Pais.append(excecao)
        lista_Cia_aerea.append(choice(df_carteira[df_carteira['País'] == excecao]['Cia Aérea'].unique()))
      else:
        lista_Pais.append(pais_var)
        lista_Cia_aerea.append(cia)
      if cont == 0:
        lista_pedidosFirmes.append(pedidos)
        cont = cont + 1
      else:
        lista_pedidosFirmes.append(0)
      ano = choice(e190)
      lista_ano.append(ano)

  elif modelo == 'EMBRAER 195 (E195)':
    while var != 0:
      entrega = randint(1,var)
      var = var - entrega
      lista_entrega.append(entrega)
      #print(var)
      lista_modelo.append(modelo)
      if pais_var == 'Clientes não divulgados':
        excecao = choice(pais_augmentation)
        lista_Pais.append(excecao)
        lista_Cia_aerea.append(choice(df_carteira[df_carteira['País'] == excecao]['Cia Aérea'].unique()))
      else:
        lista_Pais.append(pais_var)
        lista_Cia_aerea.append(cia)
      if cont == 0:
        lista_pedidosFirmes.append(pedidos)
        cont = cont + 1
      else:
        lista_pedidosFirmes.append(0)
      ano = choice(e195)
      lista_ano.append(ano)

  elif modelo == 'EMBRAER 190-E2 (E190-E2)':
    while var != 0:
      entrega = randint(1,var)
      var = var - entrega
      lista_entrega.append(entrega)
      #print(var)
      lista_modelo.append(modelo)
      if pais_var == 'Clientes não divulgados':
        excecao = choice(pais_augmentation)
        lista_Pais.append(excecao)
        lista_Cia_aerea.append(choice(df_carteira[df_carteira['País'] == excecao]['Cia Aérea'].unique()))
      else:
        lista_Pais.append(pais_var)
        lista_Cia_aerea.append(cia)
      if cont == 0:
        lista_pedidosFirmes.append(pedidos)
        cont = cont + 1
      else:
        lista_pedidosFirmes.append(0)
      ano = choice(e190_e2)
      lista_ano.append(ano)

  else:
    while var != 0:
      entrega = randint(1,var)
      var = var - entrega
      lista_entrega.append(entrega)
      #print(var)
      lista_modelo.append(modelo)
      if pais_var == 'Clientes não divulgados':
        excecao = choice(pais_augmentation)
        lista_Pais.append(excecao)
        lista_Cia_aerea.append(choice(df_carteira[df_carteira['País'] == excecao]['Cia Aérea'].unique()))
      else:
        lista_Pais.append(pais_var)
        lista_Cia_aerea.append(cia)
      if cont == 0:
        lista_pedidosFirmes.append(pedidos)
        cont = cont + 1
      else:
        lista_pedidosFirmes.append(0)
      ano = choice(e195_e2)
      lista_ano.append(ano)

  #lista_ano.append(ano)
  #print(modelo, ano)

print(np.shape(lista_entrega))
print(np.shape(lista_ano))
print(np.shape(lista_modelo))
print(np.shape(lista_Pais))
print(np.shape(lista_Cia_aerea))
print(np.shape(lista_pedidosFirmes))

dados_random_ano = {
    'Modelo do Avião': lista_modelo,
    'Quantidade entregue': lista_entrega,
    'Ano de Entrega': lista_ano,
    'País': lista_Pais,
    'Cia Aérea': lista_Cia_aerea,
    'Pedidos Firmes': lista_pedidosFirmes
}

df_random_modeloEAno_entregas = pd.DataFrame(dados_random_ano)
```
